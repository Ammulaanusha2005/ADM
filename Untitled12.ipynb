```python
import pandas as pd

df = pd.read_csv('/content/employee_data.csv')

df.head()
```

```python

df['Age'].fillna(df['Age'].median(), inplace=True)  # Impute with median for Age
df['Salary'].fillna(df['Salary'].median(), inplace=True)  # Impute with median for Salary
df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode()[0], inplace=True)  # Impute with mode for Job Satisfaction

df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])

df.head()
```

```python

df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Job_Satisfaction', 'Work_Hours_Per_Week']] = scaler.fit_transform(df[['Job_Satisfaction', 'Work_Hours_Per_Week']])

df.head()
```

```python

df['Job_Satisfaction'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()
df['Work_Hours_Per_Week'] = (df['Work_Hours_Per_Week'] - df['Work_Hours_Per_Week'].mean()) / df['Work_Hours_Per_Week'].std()

df.head()
```
