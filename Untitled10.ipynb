```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/employee_performance (1).csv')

# Check the first few rows of the dataset to understand its structure
df.head()
```

```python
# Generate a statistical summary for numerical columns
print(df.describe())

# For categorical data, you can use value counts
print(df['Department'].value_counts())
```

```python
# Encode the 'Department' column (categorical to numerical)
df['Department_Code'] = df['Department'].astype('category').cat.codes

# You can create a new column that indicates if the employee has a high performance score
df['High_Performance'] = df['Performance_Score'] >= 7
```

```python
# Group by department and calculate average salary and performance score
department_summary = df.groupby('Department').agg({
    'Monthly_Salary': 'mean',
    'Performance_Score': 'mean',
    'Years_of_Experience': 'mean'
}).reset_index()

print(department_summary)
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of Monthly Salaries
plt.figure(figsize=(8, 6))
plt.hist(df['Monthly_Salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Scatter plot of Years of Experience vs Performance Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Years_of_Experience', y='Performance_Score', data=df, color='purple')
plt.title('Scatter Plot: Years of Experience vs Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Heatmap of Numerical Features')
plt.show()
```
